```python
#Калибровка веса
weight = -0.5
input = 0.5
goal_prediction = 0.8000000000000005
step_amount = 0.01
for i in range(1000):
    prediction = input * weight
    error = (prediction - goal_prediction)**2
    print("Iteration: " + str(i) + "\tPrediction: " + str(prediction) +"\tWeight: " + str(weight))
    up_prediction = input * (weight + step_amount)
    up_error = (goal_prediction - up_prediction)**2
    down_prediction = input * (weight - step_amount)
    down_error = (goal_prediction - down_prediction)**2
    
    if down_error < up_error:
        weight -= step_amount
    if down_error > up_error:
        weight += step_amount
    if prediction == goal_prediction:
        print("Stop!")
        break
```

```python
#Вычисление направления и величины из ошибки
weight = 0.5
goal_pred = 0.8
input = 0.5
for i in range(20):
    pred = input * weight
    error = (pred - goal_pred) ** 2
    #ГРАДИЕНТНЫЙ СПУСК
    #3 эффекта
    #Первый - остановка (Если input 0, значит учиться нечему и калибровка веса будет остановлена)
    #Второй - обращение знака ( Если input отрицательный, то вес все равно будет нормально калиброваться)
    #Третий - масштабирование (Если input большой, то и смещение веса будет увеличиваться на большее число?
    #это побочный эффект)
    direction_and_amount = (pred - goal_pred)  * input
    
    weight = weight - direction_and_amount
    print("Iteration: " + str(i) + "\tPrediction: " + str(pred) +"\tWeight: " + str(weight))
    if pred == goal_pred:
        print("Stop!")
        break
```

## Пример использования α-коэффициента

```python
## При использовании чистого градиентоно спуска возникает проблема при достаточно больших входных данных
# Представим часовой маятник. Вместо того чтобы постепенно замедляться, чтобы в итоге остановиться в нулевой точке,
#а нулевая точка значит error = 0, т.е. идеальный прогноз, маятник начинает наоборот увеличивать свою амплитуду,
#удаляясь от верного прогноза.
#Чтобы такогом не просходило используют альфа коэффициент, который масштабирует прирост/убыток веса, тем самым 
#правильно его корректируя
weight, goal_pred, input = 0.1, 16, 1000
alpha = 0.000001
def neural_network(input,weight):
    prediction = input * weight
    return prediction
```

```python
#В данном примере я показал, что выбрав input в десятки раз больше я всё равно получил результат, причем очень быстро
#Однако, при еще больших значениях программа всё равно начала вести себя неуправляемо. При слишком большом  input
#Происходило переполнение, а при очень маленьком альфа движение в сторону правильного прогноза слишком сильно
#замедлялось
for i in range(50):
    pred = neural_network(input, weight)
    print(pred)
    error = (goal_pred - pred) ** 2
    delta = pred - goal_pred
    derivative = input * delta
    weight -= alpha * derivative
    if goal_pred == pred:
        print("Stop!")
        break
```

```python
weight = 0.1
input = 8.5
pred_goal = 0.8
alpha = 0.01
def neural_network(input,weight):
    prediction = input * weight
    return prediction
```

```python
for i in range(50):
    pred = neural_network(input,weight)
    print(pred)
    delta = pred - pred_goal
    derivative = input * delta
    weight -= derivative * alpha
    if pred == pred_goal:
        print("Stop!")
        break
```
