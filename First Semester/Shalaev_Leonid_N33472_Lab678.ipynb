# Лабораторная работа №6
## Условие задачи: В лифт 8-этажного дома на первом этаже вошли 5 человек. Предположим, что каждый из них с равной вероятностью может выйти на любом из этажей, начиная со второго. Найти вероятность того, что все пятеро выйдут на разных этажах.

## Решение данной задачи:


Нам оно понадобится для сравнения результата, чтобы случайно не допустить ошибку.

```python

import numpy as np
numberOfExperiment = 10000 #В начале будет столько экспериментов
probability = 0 #если вероятность будет нулевой
precision = 1e-4 #точность вычислений, которую мы хотим достигнуть в вычислениях
```

```python
while True:
    positiveResult = 0 #Счётчик положительных результатов (инкрементируется, если 5 человек вышло на разных этажах
                       #в текущем эксперименте)
        
        
    for i in range(numberOfExperiment): 
        case=np.random.choice(7,5) #Берём 5 человек и "раскидываем" случайно по этажам
        if (len(np.unique(case))==5):  #Если все 5 человек вышли на разных этажах, то засчитываем позитивный исход
            positiveResult += 1
            
    currentProb=float(positiveResult/numberOfExperiment) # находим вероятность события в numberOfExperiment случаях
    
    if np.abs(currentProb - probability) < precision: #Если мы достигли нужной точности, то ответ готов
        print("Вероятность того, что все пятеро выйдут на разных этажах: ", probability)
        break
    else:
        probability = currentProb  #иначе увеличиваем количество экспериментов в 2 раза, т.к. нужная точность  
                                   #не была достигнута
        numberOfExperiment *= 2
```

### Ответ достаточно близок к решениям на калькуляторе, всё сделано верно

# Лабораторная работа №7
## Условие задачи: Колода игральных карт содержит 52 карты, разделяющиеся на 4 различные масти по 13 карт в каждой. Предположим, что колода тщательно перетасована, так что вытаскивание любой карты одинаково вероятно. Вытащим 6 из них. Найти вероятность того, что среди этих карт не будет карт червовой масти.

## Решение данной задачи:

Ответ: 0.160258
Нам оно понадобится для сравнения результата, чтобы случайно не допустить ошибку.

```python
import numpy as np
numberOfExperiment = 10000 #В начале будет столько экспериментов
probability = 0 #если вероятность будет нулевой
precision = 1e-4 #точность вычислений, которую мы хотим достигнуть в вычислениях
```

```python
while True:
    positiveResult = 0 #Счётчик положительных результатов (инкрементируется, если среди карт не будет червовой масти
                       #в текущем эксперименте)
        
        
    for i in range(numberOfExperiment): 
        case=np.random.choice(52,6, replace=False) #Берём 6 карт. В данном случае нельзя забыть указать
                                                   #Replace=False, т.к. мы не можем вытаскивать одну и ту же карту
                                                   #из колоды несколько раз.
                
                
        #Тут нужно сделать отступление.
        #Т.к. у нас достаточно большое количество экспериментов, то мы можем сказать, что каждая 4 карта будет 
        #червовой масти. Поэтому допустим, что у каждой червовой карты есть индекс, который кратен 4.
        #Карты других мастей имеют остаток от деления 1,2 и 3, черви - 0. Получим:
        if np.prod(case%4):  #Проверяем карты на их индексы.Если все карты были НЕ червовой масти, то произведение
                            #будет >=1,  а это значит, что выражение примет вид if True и нас будет устраивать этот 
                           #результат. Иначе выражение примет вид if False, ведь всё произведение будет умножено 
                          #на ноль. Этот вариант нас не устраивает и мы переходим к следующему эксперименту.
            positiveResult += 1
            
            
    currentProb=float(positiveResult/numberOfExperiment) # находим вероятность события в numberOfExperiment случаях
    
    if (np.abs(currentProb - probability) < precision): #Если мы достигли нужной точности, то ответ готов
        print("Вероятность того, что среди шести карт не будет червей: ", probability)
        break
    else:
        probability = currentProb  #иначе увеличиваем количество экспериментов в 2 раза, т.к. нужная точность  
                                   #не была достигнута
        numberOfExperiment *= 2
```

### Ответ достаточно близок к решениям на калькуляторе, всё сделано верно

# Лабораторная работа №7
## Условие задачи:Мешалкин Сборник задач по теории вероятностей n 59. В квадрат с вершинами (0, 0), (0, 1), (1,0), (1,1) наудачу брошена точка М. Пусть (x, y) будут ее координаты. Предполагается, что вероятность попадания в область, лежащую целиком внутри квадрата, зависит лишь от площади этой области и пропорциональна ей. Найти для z = 1/4, P{max(x, y)<z}

```python

import numpy as np
numberOfExperiment = 10000 #В начале будет столько экспериментов
probability = 0 #если вероятность будет нулевой
precision = 1e-4 #точность вычислений, которую мы хотим достигнуть в вычислениях
z = 1/4
```

```python
while True:
    positiveResult = 0 #Счётчик положительных результатов (инкрементируется, если 5 человек вышло на разных этажах
                       #в текущем эксперименте)
        
        
    for i in range(numberOfExperiment): 
        x,y = np.random.random(2)
        if x < z and y < z:  #Если точка попала в нужную область, то засчитываем позитивный исход
            positiveResult += 1
            
    currentProb=float(positiveResult/numberOfExperiment) # находим вероятность события в numberOfExperiment случаях
    
    if np.abs(currentProb - probability) < precision: #Если мы достигли нужной точности, то ответ готов
        print("Вероятность того, что точка попала в квадрат 1/4 на 1/4: ", probability)
        break
    else:
        probability = currentProb  #иначе увеличиваем количество экспериментов в 2 раза, т.к. нужная точность  
                                   #не была достигнута
        numberOfExperiment *= 2
```
